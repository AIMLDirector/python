```python
#streamlit & gradio 
import matplotlib.pyplot as plt
```

```python
y = [10, 5, 8, 4, 2]
x = [5, 2, 9, 4, 7] 
plt.title("Line Graph")
plt.xlabel("Time (hr)")
plt.ylabel("Position (Km)")
plt.plot(x,y)
plt.show()
```

```python
y = [10, 5, 8, 4, 2]
x = [5, 2, 9, 4, 7] 
plt.title("Bar Graph")
plt.xlabel("Time (hr)")
plt.ylabel("Position (Km)")
plt.bar(x,y)
plt.show()
```

```python
y = [10, 5, 8, 4, 2]
x = [5, 2, 9, 4, 7] 
plt.hist(y)
plt.title("Histogram Graph")
plt.xlabel("Time (hr)")
plt.ylabel("Position (Km)")
plt.show()
```

```python
y = [10, 5, 8, 4, 2]
x = [5, 2, 9, 4, 7] 
plt.scatter(x,y)
plt.show()
```

```python
cars = ['AUDI', 'BMW', 'FORD',
        'TESLA', 'JAGUAR', 'MERCEDES']
 
data = [23, 17, 35, 29, 12, 41]
 
# Creating plot
fig = plt.figure(figsize=(10, 7))
explode = (0, 0, 0.3, 0, 0,0.4 ) 
plt.pie(data, labels=cars, explode=explode)

 
# show plot
plt.show()
```
